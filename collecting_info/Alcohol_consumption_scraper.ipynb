```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_countries_by_alcohol_consumption_per_capita"
headers = {"User-Agent": "StudentResearchBot/1.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

tables = soup.find_all("table", class_="wikitable")

# ---- Table 1 by caption ----
table1 = next(table for table in tables if table.find("caption") and "Recorded consumption of pure alcohol" in table.find("caption").get_text())

# Table 1 header (first 8 columns including Country)
header1 = [th.get_text(strip=True) for th in table1.find("tr").find_all(["th","td"])[:7]]
table1_data = [header1]

# Table 1 data
for row in table1.find_all("tr")[1:]:
    cols = row.find_all(["th","td"])
    if len(cols) >= 7:
        country = cols[0].find("a").get_text(strip=True)
        values = [col.get_text(strip=True) for col in cols[1:7]]
        table1_data.append([country]+values)

# ---- Table 2 (first and fourth columns) ----
table2 = tables[1]

def c(cell):  # clean cell
    for sup in cell.find_all("sup"): sup.decompose()
    return cell.get_text(strip=True)

header = table2.find("tr").find_all(["th","td"])
table2_data = [[c(header[0]), c(header[3])]]  # header

for row in table2.find_all("tr")[1:]:
    cols = row.find_all("td")
    if len(cols) >= 4:
        table2_data.append([c(cols[0]), c(cols[3])])
# ---- Print ----
print("TABLE 1:")
for r in table1_data: print(r)

print("\nTABLE 2:")
for r in table2_data: print(r)
```
